```python
# Q5
# 1. Input layer has 10 nodes, it also means the input is a vector with length 10
# 2. Followed by a FC (fully connected) layer with 20 nodes, activation function ReLu
# 3. Followed by a FC layer with 15 nodes, activation function ReLu
# 4. Output layer is also a FC layer with 5 nodes, activation function Sigmoid

dnn = Sequential(name="dnn")

# dense layers
dnn.add(Flatten(input_shape=(10, 10, 1)))
dnn.add(Dense(units=20, activation='relu'))
dnn.add(Dense(units=15, activation='relu'))
dnn.add(Dense(units=5, activation='sigmoid')) 

# calculate how many parameters in the DNN need to be trained.
```

```python
dnn.summary()
```

```python
# Q6
# 1. Takes a 32x32 RGB image as input
# 2. Followed by a Convolutional layer with 10 5*5 kernels, activation function relu
# 3. Followed by a 2x2 max pooling layer
# 4. Followed by a fully-connected layer with 10 nodes, activation function softmax

cnn = Sequential(name='cnn')

cnn.add(Conv2D(filters=10, kernel_size=(5, 5), activation='relu', input_shape=(32, 32, 3)))
cnn.add(MaxPool2D(pool_size=(2, 2)))
# Add Flatten before fully-connected layer to 1d vector
cnn.add(Flatten())
cnn.add(Dense(units=10, activation='softmax')) # 10 output classes

# calculate how many parameters in the CNN need to be trained.
```

```python
cnn.summary()
```
